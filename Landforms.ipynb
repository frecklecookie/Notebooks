```python
# Import modules that contain functions we need
import pandas as pd
import numpy as np
%matplotlib inline
import pylab
import matplotlib.pyplot as plt
```

### Global Mean Sea Level (GMSL) and Glacial Isostatic Adjustment (GIA)

The GMSL has changed quite a bit in the past century as seen below.

The GIA is nessesary for accurate data. This accounts for the slight enlargement of ocean basins since the end of the last glacial cycle. In this case, GIA is not cased by current glacial melt rather it is caused by the melting of ice that covered North America and Europe around 20,000 years ago.

```python
# Importing data for Global Mean Sea Level obtained from NASA
data = pd.read_csv("GMSL Data.csv")

# Sets the index to Year
gmsl = data.set_index("Year/Fraction of Year")
gmsl
```

### GMSL Variation with GIA Applied

```python
# Plots the data with year as x-axis and GMSL data as y-axis
# Data can be named
plot1 = data.plot('Year/Fraction of Year', 'GMSL  variation (GIA applied) (mm)', title = "Sea Level Height Over Time")
# Set the names of x and y axis
plot1.set(xlabel="Year", ylabel="Sea Level Variation (mm)")
```

### GMSL Comparison with GIA Applied and Not Applied

You can change the color of the lines on the graph by replacing the letter with the letter of the color you want.

Try '-r', '-m', '-y', and more.

```python
x = data['Year/Fraction of Year']
y1 = data['GMSL  variation (GIA applied) (mm)']
y2 = data['GMSL variation (GIA not applied) (mm)']

# More than one set of data can be shown on the same x-axis
# Different colors can be assigned to each data set
pylab.plot(x, y1, '-b', label='(GIA applied)')
pylab.plot(x, y2, '-c', label='(GIA not applied)')
pylab.legend(loc='upper left')
pylab.xlabel('Year')
pylab.ylabel('Sea Level Variation (mm)')
pylab.title('Sea Level Height Over Time')
```
